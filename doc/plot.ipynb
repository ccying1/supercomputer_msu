```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go

# 
data = pd.read_csv("/Users/intan/Downloads/test/task2/polus/resultseq.csv")
data_transposed = data.T

#  x  y 
shape0, shape1 = data_transposed.shape
x, y = np.linspace(-3, 3, shape0), np.linspace(0, 4, shape1)

# 
fig = go.Figure(data=[go.Surface(z=data_transposed.values, x=x, y=y)])
fig.update_layout(title="Function w(x,y) - Transposed", autosize=False, width=500, height=500)
fig.show()
```

```python
import matplotlib.pyplot as plt

# Data from the table
threads_80_90 = [2, 4, 8, 16]
acceleration_80_90 = [1.26, 1.54, 1.80, 2.29]

threads_160_180 = [4, 8, 16, 32]
acceleration_160_180 = [1.86, 2.63, 2.94, 3.49]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(threads_80_90, acceleration_80_90, marker='o', label="80 x 90")
plt.plot(threads_160_180, acceleration_160_180, marker='o', label="160 x 180")

plt.xlabel("  (Number of Threads)")
plt.ylabel(" (Acceleration)")
plt.title("Acceleration vs Number of Threads for Different Grid Sizes")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 
data = pd.read_csv("/Users/intan/Downloads/test/task2/polus/resultseq.csv")
data_transposed = data.T

#  x  y 
shape0, shape1 = data_transposed.shape
x = np.linspace(-3, 3, shape0)
y = np.linspace(0, 4, shape1)

# 
plt.figure(figsize=(8, 6))
sns.heatmap(data_transposed, xticklabels=False, yticklabels=False, cmap="magma")
plt.title("Function w(x,y) - Transposed Heatmap")
plt.show()
```
